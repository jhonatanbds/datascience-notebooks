```
from numpy import *
import matplotlib.pyplot as plt
```

```
# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))
```

```
def gradient_descent(b, m, num_iterations, learning_rate, points):
  rss = []
  b_gradient = 0
  m_gradient = 0
  for n in range(num_iterations):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m * x) + b))
        m_gradient += -(2/N) * x * (y - ((m * x) + b))
    b -= (learning_rate * b_gradient)
    m -= (learning_rate * m_gradient)
    rss.append((b_gradient, m_gradient))
    print("RSS on " + str(n) + " iteration: "+ str(rss[n]))
  plt.plot(range(1, num_iterations+1), rss)
  return b,m
```

##Execução do mesmo programa nos dados contendo anos de escolaridade versus salário, enquanto imprime o RSS a cada iteração do gradiente descendente. Plot mostrando o RSS por interação gerado a seguir.

```
points = genfromtxt('/content/gdrive/My Drive/colab_data/income.csv', delimiter=",")
learning_rate = 0.0001
num_iterations = 1000
b = 0
m = 0
b,m = gradient_descent(b,m,num_iterations,learning_rate, points)

print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(b, m, compute_error_for_line_given_points(b, m, points)))
print("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
```

## Novo algoritmo para considerar um critério de parada como um limiar de tolerância que é comparado ao tamanho do gradiente

```
def gradient_descent_thresh(b, m, learning_rate, points, thresh):
  rss = []
  b_gradient = thresh
  m_gradient = thresh
  num_iterations = 0
  while(abs(b_gradient) >= thresh and abs(m_gradient) >= thresh):
    num_iterations += 1
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m * x) + b))
        m_gradient += -(2/N) * x * (y - ((m * x) + b))
    b -= (learning_rate * b_gradient)
    m -= (learning_rate * m_gradient)
    #rss.append((b_gradient, m_gradient))
    #print("RSS on " + str(num_iterations) + " iteration: "+ str(rss[num_iterations-1]))
  #plt.plot(range(1, num_iterations+1), rss)
  return b,m
```

## Valor encontrado tal que w0 e w1 sejam aproximadamente iguais a -39 e 5 respectivamente, usando o novo algoritmo com limiar, foi com o learning rate em 0.003 e o limiar em 0.0025

```
learning_rate = 0.003
thresh = 0.0025
b = 0
m = 0
b,m = gradient_descent_thresh(b, m, learning_rate, points, thresh)
print(b,m)
```

```
def reg_simples(points):
  tmpx = 0
  tmpy = 0
  N = len(points)
  for i in range(N):
    tmpx += points[i,0]
    tmpy += points[i,1]
  mean_x = tmpx/N
  mean_y = tmpy/N
  a = 0
  b = 0
  for i in range(N):
    a += (points[i,0] - mean_x) * (points[i,1] - mean_y)
    b += (points[i,0] - mean_x)**2
  w1 = a/b
  w0 = mean_y - w1*mean_x
  return w0,w1
```
